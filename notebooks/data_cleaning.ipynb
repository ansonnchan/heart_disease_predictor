```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
df = pd.read_csv("../data/heart_disease_raw.csv")
df.head()
```

```python
df.shape
df.info()
```

```python
df.columns
```

```python
df.columns = (df.columns.str.strip().str.lower().str.replace(" ", "_"))
df.columns
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols].describe() #seem to be missing some values in some columns 
#visuaization 
df["cholesterol_level"].hist(bins=30)
plt.title("Cholesterol Level Distribution")
#doesn't seem to be any weird outliers 
```

```python
categorical_cols = df.select_dtypes(include="object").columns
for col in categorical_cols:
    print(df[col].value_counts())
    print("-" * 40)
```

```python
df["fasting_blood_sugar"].isna().mean()
```

```python
df["fasting_blood_sugar"].fillna(
    df["fasting_blood_sugar"].median(), inplace=True
)
df["fasting_blood_sugar"].isna().mean()
```

```python
yes_no_cols = [
    "smoking",
    "family_heart_disease",
    "diabetes",
    "high_blood_pressure",
    "low_hdl_cholesterol",
    "high_ldl_cholesterol",
    "heart_disease_status"
]

for col in yes_no_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0}) #binary encoding 
```

```python
ordinal_map = {"low": 0,"medium": 1,"high": 2}

ordinal_cols = [
    "exercise_habits",
    "alcohol_consumption",
    "stress_level",
    "sugar_consumption"
]

for col in ordinal_cols:
    df[col] = df[col].str.lower().map(ordinal_map)
 #ordinal coding 
```

```python
df["gender"] = df["gender"].map({"Male": 1, "Female": 0}) #gender encoding 
```

```python
df.isna().sum()
df.info()
#this shows us we have some missing values 
```

```python
#we drop alcohol_consumption column as there are many missing values (~25%)
df = df.drop(columns = ['alcohol_consumption']) 
```

```python
#i think our data set is big enough such that we can drop all missing columns without worrying our data set is too little 
df_clean = df.dropna()
print(f"Rows before: {len(df)}, Rows after dropna: {len(df_clean)}")
```

```python
df_clean.isna().sum() #should show 0 missing values 
```

```python
df_clean.to_csv("../data/cleaned_data.csv", index=False)
```
